1. This paper is about the advantages of data tidying, which is a small key component of data cleaning; it presents data tidying as a way to make the process of data cleaning easier and more effective by adopting a streamlined, consistent data structure (where each variable is a column, each observation is a row, and each type of observational unit is a table) and matching tools.
2. The tidy data standard is intended to "provide a standard way to organize data values within a data set, facilitate initial exploration and analysis of the data, and to simplify the development of data analysis
tools that work well together."
3. The first sentence is comparing families to tidy datasets. Happy families are all alike, but every unhappy family is unhappy in its own way (they all are dealing with unique problems). Similarly, tidy datasets are standardized (they link "the structure of a dataset (its physical layout) with its semantics (its meaning)," but messy datasets are all unique in the specific problems/issues that make them messy (i.e. maybe different columns are cleaner than others). The second sentence is saying that it can be simple to identify what the observations and variables are in a specific dataset, but it is much harder to provide a general definition of observations/variables that applies to all datasets because what we consider a variable or an observation depends on the context of the dataset or how we interpret the data.
4. Values are either numbers (if quantitative) or strings (if
qualitative) and are organised in two ways: every value belongs to a variable and an observation. A variable is defined as containing all values that measure the same underlying attribute across units. An observation contains all values measured on the same unit across attributes.
5. Tidy data is defined as a standard way of mapping the meaning of a dataset to its structure and has 3 main components: a) each variable forms a column, b) each observation forms a row, and c) each type of observational unit forms a table.
6. The 5 most common problems with messy datasets are: 1) column headers are values, not variable names, 2) multiple variables are stored in one column, 3) variables are stored in both rows and columns, 4) multiple types of observational units are stored in the same table, 5) a single observational unit is stored in multiple tables. The data in table 4 are messy because the column headers are values (<$10k, etc.) instead of variable names and the variables are stored in both rows and columns (income bracket in the columns and religion in the rows). Melting a dataset involves turning columns into rows. It is parameterised by a list of columns that are already variables (colvars). The other columns are converted into: a) a new variable column that contains repeated column headings and b) a new variable value that contains the concatenated data values from the previously separate columns. A molten dataset results from melting.
7. Data table 11 is messy because it has column headers that are values (d1, d2, etc), multiple variables are stored in the same column (element stores tmax and tmin), and it contains a lot of missing data. Data table 12 is molten/tidy because each observation forms a row (each row represents a single specific data with tmax or tmin value), each variable forms a column (variables are separated), and each type of observational unit forms a table (all the data is organized into a single table).
8. The chicken and egg problem Wickham describes is the linkage that forms between the tidy tools and the tidy data if tidy data is only as useful as the tools that work with it, making it easy to get stuck in a local maxima where independently changing data structures or data tools will not improve workflow. He hopes that others build on the "tidy data" framework to develop better data storage strategies and tools (i.e. using methodologies from fields like user-testing to improve understanding of the cognitive side of data analysis).

```python
import pandas as pd
import numpy as np
import seaborn as sns
df = pd.read_csv('/content/sample_data/sharks.csv', low_memory=False)
print( df.shape, '\n')
df.head()
```

```python
df['Type'].value_counts()
```

```python
type = df['Type'] # Create a temporary Type vector
type = type.replace(['Sea Disaster', 'Boat','Boating','Boatomg'],'Watercraft') # Consolidate all watercraft/boating values
type.value_counts()

type = type.replace(['Invalid', 'Questionable','Unconfirmed','Unverified','Under investigation'],np.nan) # Consolidate all "unclear" values, replace with missing value
type.value_counts()

df['Type'] = type # replace 'type' variable with the cleaned version
del type # destroy the temporary vector

df['Type'].value_counts()
```

```python
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace(['UNKNOWN', 'F','M','2017'],np.nan) # handle missing values
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace('y','Y') # handle messy values
pd.crosstab(df['Type'],df['Fatal (Y/N)'],normalize='index')
```
